# Cleaning the dataset

### Reading data set, and collecting types of values in columns.

```python
import pandas as pd
df = pd.read_csv('Fire_Incidents.csv')
df.head()
col = df.columns
dtypes = []
for i in col:
    dtypes.append(df[i].dtype)
```

### Displaying dataset

```python
df
```

### Checking unique values of City column, and renaming it to "San Francisco"

```python
df['City'].unique()
```

```python
df = df.replace('SF', 'San Francisco')
df = df.replace('SAN FRANCISCO', 'San Francisco')
```

```python
df['City'].unique()
```

### Deleting columns where null is over 90% 

```python
print(round(100*(df.isnull().sum()/len(df)),2)[:32])
print(round(100*(df.isnull().sum()/len(df)),2)[32:64])
```

```python
for i in df.columns:
    if round(100*(df.isnull().sum()/len(df)),2)[i] > 90:
        df.drop(i, axis=1, inplace=True)
```

### Replacing null values with the mean of the remaining values for each column

```python
round(100*(df.isnull().sum()/len(df)),2)
```

```python
from sklearn.impute import SimpleImputer
numeric_df = df.select_dtypes(exclude='object')
for col in numeric_df.columns:
    df[col].fillna(float(df[col].mean()), inplace = True)
```

```python
round(100*(df.isnull().sum()/len(df)),2)
```

```python
for i in df.columns:
    if round(100*(df.isnull().sum()/len(df)),2)[i] > 30:
        df.drop(i, axis=1, inplace=True)
```

```python
round(100*(df.isnull().sum()/len(df)),2)
```

## Deleting unnecessary columns

```python
list = ['Call Number','Arrival DtTm','Close DtTm','Battalion','Suppression Units','Suppression Personnel','EMS Units','EMS Personnel','Other Units','Other Personnel','Fire Fatalities','Fire Injuries','Primary Situation','Mutual Aid','Action Taken Primary','Property Use']
for i in list:
    df.drop(i, axis=1, inplace=True)
```

## Deleting rows with null values

```python
round(100*(df.isnull().sum()/len(df)),2)
```

```python
for col in df.columns:
    df = df[df[col].notna()]
```

```python
round(100*(df.isnull().sum()/len(df)),2)
```
